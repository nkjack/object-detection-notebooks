- architecture:
    - 5 conv layers with some max pool
    - 3 fully connected layers with final softmax
    - to reduce overfitting they used dropout on the fully connected layers
- error rate ? what is it and how it is measured

### Notes
- Not implemented here:
    - Data augmentation:
        - 224 x 224 patches from original 256 x 256 which increased the size of the training size by a factor of 2048 in the original papaer.
        - PCA on the set of RGB pixel values.
- Dropout - "Dropout roughly doubles the number of iterations required to converge."
    - I removed the dropout to overfit easily on my sample training.
- Training
    - SGD
        - b_size = 128
        - momentum = 0.9
        - weight_decay = 0.0005 (important for the model to learn)
    - Initialization
        - Weights with 0-mean Gaussian Distribution and 0.01 std
        - Biases in 2,4,5 and fc layers to 1. ( accelerated early stage of learning)
    - Learning rate
        - Initialized to 0.01 and reduced three times prior to termination.
        - "The heuristic which we followed was to divide the learning rate by 10 when the validation error rate stopped improving with the current learning rate."
- Evaluation
    - what is top-1 / top-5 error-rate - [stackoverflow](https://stats.stackexchange.com/questions/156471/imagenet-what-is-top-1-and-top-5-error-rate)

```python
import os
import re
import sys
import glob
import numpy as np
import matplotlib.pyplot as plt
import unittest
import torch
import torchvision
import torchvision.transforms as T

%matplotlib inline
%load_ext autoreload
%autoreload 2

seed = 42
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
plt.rcParams.update({'font.size': 12})
test = unittest.TestCase()
```

```python
data_dir = os.path.expanduser('~/.pytorch-datasets')

# resizing only to fit sizes of original paper
transform = T.Compose([
    T.Resize((224,224)),
    T.ToTensor()
])

# ds_train = torchvision.datasets.ImageNet(root=data_dir, download=True, train=True, transform=tvtf.ToTensor())
ds_train = torchvision.datasets.CIFAR10(root=data_dir, download=True, train=True, transform=transform)
ds_test = torchvision.datasets.CIFAR10(root=data_dir, download=True, train=False, transform=transform)

print(f'Train: {len(ds_train)} samples')
print(f'Test: {len(ds_test)} samples')

x0,_ = ds_train[0]
in_size = x0.shape
num_classes = 10
print('input image size =', in_size)
```

```python
import utils.training as training
import alexnet
torch.manual_seed(seed)

# Define a tiny part of the CIFAR-10 dataset to overfit it
batch_size = 128
max_batches = 25
dl_train = torch.utils.data.DataLoader(ds_train, batch_size, shuffle=False)
dl_test = torch.utils.data.DataLoader(ds_test, batch_size, shuffle=False)

# Create model, loss and optimizer instances
model = alexnet.AlexNet(in_size=(3,224,224), out_classes=10)
```

```python
# weights initialization
model = model.apply(alexnet.weights_init)
conv_2_4_5 = [4,11,14]
for i in conv_2_4_5:
    model.feature_extractor[i].bias.data.fill_(1)
```

```python
# print amount of parameters
pytorch_total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"{pytorch_total_params:,}") 
```

```python
from utils.train_results import FitResult

loss_fn = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9, weight_decay=0.0005)

# Use TorchTrainer to run only the training loop a few times.
trainer = training.TorchTrainer(model, loss_fn, optimizer, device)
best_acc = 0
# for i in range(30):
#     res = trainer.train_epoch(dl_train, max_batches=max_batches, verbose=(i%2==0))
#     best_acc = res.accuracy if res.accuracy > best_acc else best_acc
    
    
fit_res = trainer.fit(dl_train, dl_test, num_epochs=30, checkpoints=None, early_stopping=10, max_batches=10)
epo, loss_tr, acc_tr, loss_te, acc_te = fit_res
fit_res = FitResult(epo, [loss.item() for loss in loss_tr], acc_tr, [loss.item() for loss in loss_te], acc_te)
```
