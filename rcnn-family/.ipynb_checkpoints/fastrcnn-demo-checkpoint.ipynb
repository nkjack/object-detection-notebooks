```python
from dataset import PennFudanDataset
import torchvision.transforms as T

im_size = 64
tf = T.Compose([
    # Resize to constant spatial dimensions
#     T.Resize((im_size, im_size)),
    # PIL.Image -> torch.Tensor
    T.ToTensor(),
    # Dynamic range [0,1] -> [-1, 1]
#     T.Normalize(mean=(.5,.5,.5), std=(.5,.5,.5)),
])

ds = PennFudanDataset(dataset_dir,  tf)
```

```python
ds[0]
```

```python
import torchvision
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
 
# load a model pre-trained pre-trained on COCO
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
```

```python
model = model.to(device)
```

```python
# pick one image from the test set
img, _ = ds[5]
# img = T.ToTensor()(img)
# put the model in evaluation mode
model.eval()
with torch.no_grad():
    prediction = model([img.to(device)])
```

```python
prediction
```

```python
from torchvision.utils import make_grid
from PIL import Image, ImageDraw

def draw_bbox(tensor, bboxes, fill=None, outline=None, width=1,):    
    im = T.ToPILImage()(tensor.cpu())
    for bbox in bboxes.cpu():
        ImageDraw.Draw(im).rectangle(np.array(bbox), fill=fill, outline=outline, width=width)
    return im
```

```python
bbox = prediction[0]['boxes'][:5]
draw_bbox(img, bbox, width=2)
```

```python
my_embedding = torch.zeros(1000, 1024)
def hook(module, inputs, outputs): 
#     print(inputs[0].size(), outputs)
    my_embedding.copy_(inputs[0].cpu())

h = model.roi_heads.box_predictor.bbox_pred.register_forward_hook(hook)

img, _ = ds[5]
model.eval()
# with torch.no_grad():
prediction = model([img.to(device)])
    
h.remove()
```
